```python
def get_broad_category(sub_category):
    if sub_category == 'Pop':
         return "Pop"
    if sub_category == 'edm':
         return "EDM"
    if sub_category == 'hip hop':
         return "Hip-Hop"
    if sub_category == 'dance pop':
         return "Pop"
    if sub_category == 'rap':
         return "Hip-Hop"
    if sub_category == 'canadian pop':
         return "Pop"
    if sub_category == 'modern rock':
         return "Rock"
    if sub_category == 'miami hip hop':
         return "Hip-Hop"
    if sub_category == 'r&b':
         return "R&B"
    if sub_category == 'post-teen pop':
         return "Pop"
    if sub_category == 'uk pop':
         return "Pop"
    if sub_category == 'metropopolis':
         return "Pop"
    if sub_category == 'singer-songwriter pop':
         return "Pop"
    if sub_category == 'country':
         return "Country"
    if sub_category == 'melodic rap':
         return "Hip-Hop"
```
